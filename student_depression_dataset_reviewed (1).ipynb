# **Data** Cleaning & Preprocessing

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

## Exploratory Data Analysis (EDA)


```python
df['Financial Stress'].value_counts().plot(kind='bar', figsize=(6,4))
plt.title('Distribution of Financial Stress')
plt.show()
```

A higher proportion of students with financial stress may show symptoms of depression.

```python
df['Degree'].value_counts().plot(kind='bar', figsize=(10,6))
plt.title('Distribution of Degrees')
plt.show()
```

Degree distribution shows which academic groups are more represented.

```python
df['Depression'].value_counts().plot(kind='bar', figsize=(6,4))
plt.title('Depression Levels Among Students')
plt.show()
```

The plot highlights the prevalence of depression among students.

```python
sns.boxplot(x=df['Gender'], y=df['Depression'])
plt.title('Depression by Gender')
plt.show()
```

it can be seen that Gender doesn't pay any significant role in depressed students

```python
top_cities = df.City.value_counts().head(10)
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Top 10 Cities")
sns.barplot(x=top_cities.index, y=top_cities);

for i in range(len(top_cities)):
    plt.text(i, top_cities[i], top_cities[i], ha='center', va='bottom')
```

The graph highlights the top 10 cities with most no. of cases

```python
sns.countplot(x='Family History of Mental Illness', hue='Depression', data=df, palette='coolwarm')
plt.title('Depression by Family History of Mental Illness')
plt.xlabel('Family History of Mental Illness (Yes/No)')
plt.ylabel('Count')
plt.legend(title='Depression', loc='lower right')

for p in plt.gca().patches:
    plt.text(p.get_x() + p.get_width()/2., p.get_height(),
             '{:1.0f}'.format(p.get_height()),
             ha='center', va='bottom')
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Features with higher correlation to `Depression` are potential predictors.

## Conclusion
